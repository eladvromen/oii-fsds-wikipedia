```python
from wordcloud import WordCloud
from datetime import timedelta
from collections import Counter
import matplotlib.pyplot as plt
import nltk
from nltk.corpus import stopwords
import string
import math

# Ensure NLTK stopwords are downloaded
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))

def preprocess_text(text, n=1):
    """
    Preprocess the text by lowercasing, removing stopwords and punctuation,
    and creating n-grams.
    """
    tokens = text.lower().split()
    tokens = [word.strip(string.punctuation) for word in tokens if word not in stop_words and word.isalpha()]
    
    # Generate n-grams
    if n > 1:
        ngrams = zip(*[tokens[i:] for i in range(n)])
        tokens = [' '.join(ngram) for ngram in ngrams]
    
    return tokens

def generate_word_clouds(data, start_date=None, end_date=None, division=3, interval_type=None, n=1):
    """
    Generate word clouds for textual data by time intervals with detailed plot titles.

    Parameters:
    - data: DataFrame with 'text', 'timestamp', 'year', and 'month' columns.
    - start_date: str, the starting date for the word cloud analysis (default: first date in data).
    - end_date: str, the ending date for the word cloud analysis (default: last date in data).
    - division: int, number of equal time segments to divide the data (ignored if interval_type is provided).
    - interval_type: str, 'year' or 'month' to generate word clouds for each unique year or month.
    - n: int, the n-gram size for processing text (1 for unigrams, 2 for bigrams, etc.).
    """
    # Convert timestamps to datetime
    data['timestamp'] = pd.to_datetime(data['timestamp'])

    # Parse and localize start_date and end_date to match timestamp's timezone-awareness
    start_date = pd.to_datetime(start_date).tz_localize(data['timestamp'].dt.tz) if start_date else data['timestamp'].min()
    end_date = pd.to_datetime(end_date).tz_localize(data['timestamp'].dt.tz) if end_date else data['timestamp'].max()
    
    # Filter data within the date range
    data = data[(data['timestamp'] >= start_date) & (data['timestamp'] <= end_date)]

    # Determine intervals based on interval_type
    if interval_type == 'year':
        unique_intervals = sorted(data['year'].unique())
        interval_col = 'year'
    elif interval_type == 'month':
        unique_intervals = sorted(data['month'].unique())
        interval_col = 'month'
    else:
        # For division-based splitting
        total_duration = (end_date - start_date) / division
        unique_intervals = [(start_date + i * total_duration, start_date + (i + 1) * total_duration) for i in range(division)]
        interval_col = None

    # Determine grid layout for plots
    num_intervals = len(unique_intervals)
    cols = 3  # Number of columns for the grid
    rows = math.ceil(num_intervals / cols)

    fig, axes = plt.subplots(rows, cols, figsize=(cols * 5, rows * 4))
    axes = axes.flatten()  # Flatten to iterate easily even if single row

    # Generate word clouds for each interval
    for idx, interval in enumerate(unique_intervals):
        # Filter data within interval
        if interval_col == 'year':
            interval_data = data[data[interval_col] == interval]
            interval_label = f"Year: {interval}"
        elif interval_col == 'month':
            interval_data = data[data[interval_col] == interval]
            interval_label = f"Month: {interval}"
        else:
            interval_start, interval_end = interval
            interval_data = data[(data['timestamp'] >= interval_start) & (data['timestamp'] < interval_end)]
            interval_label = f"From {interval_start.date()} to {interval_end.date()}"

        # Concatenate all text within the interval
        text_data = ' '.join(interval_data['text'].dropna())

        # Check for empty text
        if text_data.strip():
            # Process text and create word cloud
            words = preprocess_text(text_data, n=n)
            word_freq = Counter(words)
            wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(word_freq)

            # Plot word cloud
            axes[idx].imshow(wordcloud, interpolation='bilinear')
            axes[idx].axis('off')
            axes[idx].set_title(f'{n}-gram Word Cloud for {interval_label}')
        else:
            # No text data for the interval
            axes[idx].text(0.5, 0.5, 'No Data', ha='center', va='center', fontsize=16, color='gray')
            axes[idx].axis('off')
            axes[idx].set_title(f'{n}-gram Word Cloud for {interval_label}')

    # Hide any unused subplots
    for j in range(idx + 1, len(axes)):
        axes[j].axis('off')

    plt.tight_layout()
    plt.show()

# Example call (assuming columns 'year' and 'month' already exist in the data)
generate_word_clouds(data, n=3 ,start_date="2014-01-01", end_date="2024-01-01",interval_type='year')  # Generates bigram word clouds

```

```python
def plot_top_ngrams(data, start_date=None, end_date=None, division=3, interval_type=None, n=1, top_k=10):
    """
    Plot a histogram of the most frequent words or n-grams in the textual data.

    Parameters:
    - data: DataFrame with 'text', 'timestamp', 'year', and 'month' columns.
    - start_date: str, the starting date for the histogram (default: first date in data).
    - end_date: str, the ending date for the histogram (default: last date in data).
    - division: int, number of equal time segments to divide the data (ignored if interval_type is provided).
    - interval_type: str, 'year' or 'month' to generate histograms for each unique year or month.
    - n: int, the n-gram size (1 for unigrams, 2 for bigrams, etc.).
    - top_k: int, the number of top n-grams to display in the histogram.
    """
    # Convert timestamps to datetime
    data['timestamp'] = pd.to_datetime(data['timestamp'])

    # Parse and localize start_date and end_date to match timestamp's timezone-awareness
    start_date = pd.to_datetime(start_date).tz_localize(data['timestamp'].dt.tz) if start_date else data['timestamp'].min()
    end_date = pd.to_datetime(end_date).tz_localize(data['timestamp'].dt.tz) if end_date else data['timestamp'].max()
    
    # Filter data within the date range
    data = data[(data['timestamp'] >= start_date) & (data['timestamp'] <= end_date)]

    # Determine intervals based on interval_type
    if interval_type == 'year':
        unique_intervals = sorted(data['year'].unique())
        interval_col = 'year'
    elif interval_type == 'month':
        unique_intervals = sorted(data['month'].unique())
        interval_col = 'month'
    else:
        # For division-based splitting
        total_duration = (end_date - start_date) / division
        unique_intervals = [(start_date + i * total_duration, start_date + (i + 1) * total_duration) for i in range(division)]
        interval_col = None

    # Create subplots for histograms in a grid layout
    num_intervals = len(unique_intervals)
    cols = 3  # Number of columns in the grid
    rows = math.ceil(num_intervals / cols)
    
    fig, axes = plt.subplots(rows, cols, figsize=(cols * 5, rows * 4))
    axes = axes.flatten()

    # Generate histograms for each interval
    for idx, interval in enumerate(unique_intervals):
        # Filter data within the interval
        if interval_col == 'year':
            interval_data = data[data[interval_col] == interval]
            interval_label = f"Year: {interval}"
        elif interval_col == 'month':
            interval_data = data[data[interval_col] == interval]
            interval_label = f"Month: {interval}"
        else:
            interval_start, interval_end = interval
            interval_data = data[(data['timestamp'] >= interval_start) & (data['timestamp'] < interval_end)]
            interval_label = f"From {interval_start.date()} to {interval_end.date()}"

        # Concatenate all text within the interval
        text_data = ' '.join(interval_data['text'].dropna())

        # Check for empty text
        if text_data.strip():
            # Process text to extract n-grams and count frequencies
            words = preprocess_text(text_data, n=n)
            word_freq = Counter(words)
            top_ngrams = word_freq.most_common(top_k)
            ngrams, counts = zip(*top_ngrams) if top_ngrams else ([], [])

            # Plot histogram
            axes[idx].barh(ngrams, counts, color='skyblue')
            axes[idx].set_xlabel('Frequency')
            axes[idx].set_title(f'Top {top_k} {n}-grams for {interval_label}')
            axes[idx].invert_yaxis()
        else:
            # No text data for the interval
            axes[idx].text(0.5, 0.5, 'No Data', ha='center', va='center', fontsize=16, color='gray')
            axes[idx].axis('off')
            axes[idx].set_title(f'{n}-gram Histogram for {interval_label}')

    # Hide any unused subplots
    for j in range(idx + 1, len(axes)):
        axes[j].axis('off')

    plt.tight_layout()
    plt.show()

# Example call (assuming columns 'year' and 'month' already exist in the data)
plot_top_ngrams(data, interval_type='year', n=3, top_k=10)

```
